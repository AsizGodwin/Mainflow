```python
#1. how many people have heart disease and how many people doesnt have heart disease?df.value_counts())
df.value_counts()
```

```python
#bar chart
df.value_counts().plot(kind='bar',color=["orchid","salmon"])
plt.title("heart disease values")
plt.xlabel("1=heart dsease,0=No heart disease")
plt.ylabel("amount");
```

```python
#pie chart
df.value_counts().plot(kind='pie',figsize=(8,6))
plt.legend(["disease","no disease"]);
```
