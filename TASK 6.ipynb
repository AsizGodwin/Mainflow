```python
#1 type of shows
df["type"].value_counts()
```

```python
df.type.value_counts().plot(kind='bar', figsize=(6,5), color=('red', 'green', 'blue', 'yellow', 'orange', 'violet', 'seagreen', 'salmon'))
plt.xlabel('type')
plt.ylabel("type of movies")
plt.title("Number of movie types")
plt.show()
```

```python
#2. No. of show realse year
df.release_year.value_counts().plot(kind='bar', figsize=(26,10))
plt.xlabel('release year')
plt.ylabel("Number of show")
plt.title("year")
plt.xticks(rotation=90)
plt.show()
```

```python
#3. Total no of unique
num_genres = df['listed_in'].nunique()
print("Number of unique genres:", num_genres)
```

```python
df.listed_in.value_counts()
```

```python
df.listed_in.value_counts().plot(kind='line', figsize=(32,9))
plt.show()
```

```python
#4. Release by runtime
df.groupby('release_year')['duration'].value_counts()
```

```python
df.groupby('release_year')['duration'].nunique().plot(kind='bar', figsize=(32,10))
plt.title("Movies of varying runtime Vs year")
plt.xlabel("Release_year")
plt.ylabel("Count of movies with different runtime")
plt.show()
```
