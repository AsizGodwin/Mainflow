```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("householdtask3.csv")
df
```

```python
#visualising data using pandas data frame
pd.DataFrame(df)
```

```python
#bar chart
plt.figure(figsize=(10,7))
plt.bar(df["year"],df["expenditure"],color="blue",label="household")
plt.xlabel("year")
plt.ylabel("expence")
plt.title("household expenses")
plt.legend()
plt.show
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["year"],df["income"],marker="o",color="blue",label="household")
plt.xlabel("income")
plt.ylabel("income")
plt.legend()
plt.show
```
